```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression #binary classification we use Logistic reg
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head() #class 0 is a legit transaction and class 1 is fraudulant
```

```python
#check if we have missing value
credit_card_data.isnull().sum()
```

```python
#distribution of legit transaction and fradulant transaction
credit_card_data['Class'].value_counts()  #data is very unbalanced
```

```python
#before processing the data, we need to analyze the data
#separating the data for analysis
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
#statistical method of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()   #mean of fraud is much more, its an important distinction
```

```python
credit_card_data.groupby('Class').mean()
```

```python
#Under-sampling
#build a sample dataset containing similar distribution of normal transaction and fradulant transaction
#the dataset is very unbalanced, the fraudulant data is only 492 so we'll also take 492 random value from legt dataset
legit_sample=legit.sample(n=492)
```

```python
#concatinating the two data set of legit and fraudulant dataset
new_dataset=pd.concat([legit_sample,fraud],axis=0)  #we want to add fraud below legit sample row-wise
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean() #to ckeck if we got good sample, if bad sample then the means would be very diff from
                                    #each other but here they are close here so good sample
```

```python
#splitting the data into features and targets
X=new_dataset.drop(columns='Class',axis=0)
Y=new_dataset['Class']
```

```python
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model=LogisticRegression() #means we are loading one instance of Logistic reg
```

```python
#trainingg the model with tarining data
model.fit(X_train,Y_train)
```

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print("Accuracy on training data : ",training_data_accuracy)
```

```python
#accuracy on testing data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("Accuracy on testing data : ",testing_data_accuracy)
```
